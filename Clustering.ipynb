```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans

data = pd.read_csv('Table - Лист1.csv')
data.head()
```

```python
data = data.iloc[1:, 3:29]
data.replace(',', '.', inplace=True)
data = data.fillna(0)
columns_to_use = ['Unnamed: 10', 'Unnamed: 13', 'Unnamed: 24', 'Unnamed: 28', 'Pandas', 'kNN']
data[columns_to_use] = data[columns_to_use].apply(pd.to_numeric, errors='coerce').fillna(0)

data.head()
```

```python
data['Unnamed: 28'] *= 10
data['kNN'] *= 5
data['Pandas'] *= 5

data.head()
```

```python
data['Total Score'] = data[columns_to_use].sum(axis=1)

X = data[['Total Score']].values
```

```python
kmeans = KMeans(n_clusters=3).fit(X)

data['Cluster'] = kmeans.labels_
```
